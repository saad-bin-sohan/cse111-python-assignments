```python
#task 5 done

class Employee:

    def __init__(self, employee_name, working_period):
        self.workingPeriod = working_period
        self.name = employee_name

    @classmethod
    def employeeByJoiningYear(cls, name_, join_period):
        return cls( name_, (2023 - join_period ) )
    
    @staticmethod
    def experienceCheck(dur, gender):
        if gender == 'male':
            result = 'He is '
        else:
            result = 'She is '
        if dur < 3:
            result += 'not experienced'
        else:
            result += 'experienced'

        return result
    

# Write your code here
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))

# Output
# 3
# 7
# Dororo
# Harry
# He is not experienced
# She is experienced
```

```python
#TASK 6 DONE
class Laptop:
    laptopCount = 0
    def __init__(self, brand, count):
        self.name = brand
        self.count = count
        Laptop.laptopCount += count

    @staticmethod
    def advantage():
        print("Laptops are portable")

    @staticmethod
    def resetCount():
        Laptop.laptopCount = 0

# Write your code here

lenovo = Laptop("Lenovo", 5)
dell = Laptop("Dell", 7)
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)

# Output
# Lenovo 5
# Dell 7
# Total number of Laptops 12
# Laptops are portable
# Total number of Laptops 0
```

```python
#task 7 done

class Cat:

  Number_of_cats = 0

  @classmethod
  def no_parameter(cls):
    clr = "White"
    act = "sitting"
    return cls(clr, act)
  @classmethod
  def first_parameter(cls, color):
    clr = color
    act = "sitting"
    return cls(clr, act)
  @classmethod
  def second_parameter(cls, action):
    clr = "Grey"
    act = action
    return cls(clr, act)

  def __init__(self, color, action):
    self.color = color
    self.action = action
    Cat.Number_of_cats += 1

  def printCat(self):
    print(f"{self.color} cat is {self.action}")

  def changeColor(self, clr):
    self.color = clr

# Write your code here
    
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task 8 done
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, rad, height):
        self.radius, self.height = rad, height
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        Cylinder.radius, Cylinder.height = rad, height
        print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")

    @classmethod
    def swap(cls, h, r):
        return cls(r, h)

    @classmethod
    def changeFormat(cls, str_f):
        r, h = str_f.split("-")
        r, h = float(r), float(h)
        return cls(r, h)

    @staticmethod
    def area(r, h):
        area = 2 * math.pi * r * ( r + h )
        print(f"Area: {area}")

    @staticmethod
    def volume(r, h):
        volume = math.pi * (r ** 2) * h
        print(f"Volume: {volume}")

# Write your code here

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)

# output
# Default radius=5 and height=18.
# Updated: radius=0 and height=0.
# Area:  0.0
# Volume:  0.0
# ===============================
# Default radius=0 and height=0.
# Updated: radius=3 and height=8.
# Area:  207.34511513692635
# Volume:  226.194671058465
# 1===============================
# Default radius=3 and height=8.
# Updated: radius=7.0 and height=13.0.
# Area:  879.645943005142
# Volume:  2001.1945203366981
# ===============================
# Default radius=7.0 and height=13.0.
# Updated: radius=0.3 and height=5.56.
# Area:  11.045839770021713===============================
# Default radius=0.3 and height=5.56.
# Updated: radius=3 and height=5.
# Volume:  141.3716694115407
```

```python
#task 10 done

class SultansDine:
    branch_no=0
    total_sell=0
    branch_list=[]
    def __init__(self,branch):
        self.branch=branch
        SultansDine.branch_no+=1
    def sellQuantity(self,quantity):
        self.sellQuantity=quantity
        if self.sellQuantity < 10:
            self.branch_sell = quantity * 300
        elif self.sellQuantity < 20:
            self.branch_sell = quantity * 350
        else:
            self.branch_sell = quantity * 400
        SultansDine.total_sell+=self.branch_sell
        SultansDine.branch_list.append(self.branch)
        SultansDine.branch_list.append(self.branch_sell)
    def branchInformation(self):
        print(f'Branch Name: {self.branch}\nBranch Sell:{self.branch_sell} Taka')
    @classmethod
    def details(cls):
        print(f'Total Number of branch(s): {SultansDine.branch_no}\nTotal Sell: {SultansDine.total_sell} Taka')
        for i in range(0,len(SultansDine.branch_list),2):
            print(f"Branch Name: {SultansDine.branch_list[i]}, Branch Sell: {SultansDine.branch_list[i+1]} Taka")
            print(f"Branch consists of total sell's: {round((SultansDine.branch_list[i+1]/SultansDine.total_sell)*100,2)}%")

# Write your code here

SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
