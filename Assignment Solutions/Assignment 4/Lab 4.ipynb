```python
#task 1 done

class Customer:

  def __init__(self, name):
    self.name = name

  def greet(self, nam = None):
    if nam is not None:
      print(f"Hello {nam}!")
    else:
      print("Hello")

  def purchase(self, *args):
    print(f"{self.name}, you purchased {len(args)} item(s):")
    for elem in args:
      print(elem)

#driver code
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task 2 done

class Panda:

  def __init__(self, name, gender, age):
    self.name = name
    self.gender = gender
    self.age = age

  def food_manager(self, time):

    try:
      if 3 <= time <= 5:
        return 'Mixed Veggies'

      elif 6<= time <= 8:
        return 'Eggplant & Tofu'

      elif  9<= time <= 11:
        return 'Broccoli Chicken'

    except:
      return 'bamboo leaves'

  def sleep(self, time = None):

    result = self.food_manager(time)

    if result != 'bamboo leaves':
      return f"{self.name} sleeps {time} hours daily and should have {result}"
    else:
      return f"{self.name}'s duration is unknown thus should have only bamboo leaves"

#driver code

panda1 = Panda("Kunfu", "Male", 5)
panda2 = Panda("Pan Pan", "Female",3)
panda3 = Panda("Ming Ming", "Female",8)

print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))

print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))

print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#task 7 done

class Match:
  print("5..4..3..2..1.. Play !!!")

  def __init__(self, teams):
    self.run = 0
    self.over = 0
    self.wicket = 0
    self.batting, self.bowling = teams.split('-')

  def add_run(self, run):
    self.run += run
  def add_over(self, over):
    if over >= 5:
      print(f"Warning! Cannot add {over} over/s (5 over match)")
    else:
      self.over += over
  def add_wicket(self, wicket):
    self.wicket += wicket

  def print_scoreboard(self):
    return f"Batting Team: {self.batting}\nBowling Team: {self.bowling}\nTotal runs: {self.run} Wickets: {self.wicket} Overs: {self.over}"

# Write your code here.
match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
#task 9 done

class Batsman:

  def __init__(self, *args):
    
    if len(args) == 2:
      self.name = 'New Batsman'
      self.runs_scored = args[0]
      self.balls_faced = args[1]

    elif len(args) ==3:
      self.name = args[0]
      self.runs_scored = args[1]
      self.balls_faced = args[2]

  
  def printCareerStatistics(self):
    print(f"Name: {self.name}\nRuns Scored: {self.runs_scored} , Balls Faced: {self.balls_faced}")

  def setName(self, name):
    self.name = name

  def battingStrikeRate(self):
    return ( (self.runs_scored / self.balls_faced) *100 )

# Write your code here
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#task 12 done

class TaxiLagbe:

  def __init__(self, taxi_number, area):
    self.taxi_number = taxi_number
    self.area = area
    self.all_passenger_list = []
    self.total_passenger_count = 0
  
  def addPassenger(self, *args):
    self.total_passenger_count += 1

    if self.total_passenger_count>=4:
      print("Taxi Full! No more passengers can be added.")
    
    elif self.total_passenger_count <4:
      for passerngerName_fare in args:
        passenger_name, fare = passerngerName_fare.split('_')
        self.all_passenger_list.append(  ( passenger_name, int(fare) ) )
        print(f"Dear {passenger_name}! Welcome to TaxiLagbe.")
    
  def printDetails(self):
    self.fare = 0

    print(f"Trip info for Taxi number: {self.taxi_number}\nThis taxi can cover only {self.area} area.\nTotal passengers: {len(self.all_passenger_list)}\nPassenger lists:")

    for i in range(len(self.all_passenger_list)):
      self.fare += self.all_passenger_list[i][1]

      if i == len(self.all_passenger_list)-1:
        print(f"{self.all_passenger_list[i][0]}")

      else:
        print(f"{self.all_passenger_list[i][0]}, ", end='')

    print(f"Total collected fare: {self.fare} Taka")

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#task 13 done

class Account:

  def __init__(self, name = 'Default Account', balance=0.0):
    self.name = name
    self.balance = balance

  def details(self):
    return f"{self.name}\n{self.balance}"

  def withdraw(self, amount_to_be_withdrawn):

    if  self.balance - amount_to_be_withdrawn <= 3070:
      print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")

    else:
      self.balance -= amount_to_be_withdrawn
      print(f"Withdraw successful! New balance is: {self.balance}")

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
#task 14 done

class StudentDatabase:
  def __init__(self, name, id):
    self.name = name
    self.id = id
    self.grades = {}
    self.each_semester_grade = {}
  
  def calculateGPA(self, course_list, semester):
    all_course_list = []
    gpa_this_sem_only = 0
    avg_gpa_this_sem = 0
    for course_grade in course_list:
      course_title, grade = course_grade.split(': ')
      all_course_list.append(course_title)
      gpa_this_sem_only += float(grade)
    avg_gpa_this_sem = gpa_this_sem_only / len(course_list)
    avg_gpa_this_sem = round(avg_gpa_this_sem, 2)

    temp = {tuple(all_course_list) : avg_gpa_this_sem}
    self.grades[semester] = temp
    self.each_semester_grade[semester] = avg_gpa_this_sem
  def printDetails(self):
    print(f"Name: {self.name}\nID: {self.id}")
    
    for semester, courseTuple_grade_dict in self.grades.items():
      print(f"Courses taken in {semester} :")

      for course_tuple in courseTuple_grade_dict.keys():
        for course in course_tuple:
          print(f"{course}")
      print(f"GPA: {self.each_semester_grade[semester]}")

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
