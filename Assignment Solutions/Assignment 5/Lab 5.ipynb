```python
#task 2 done

class Triangle:
  def __init__(self, base, height):
    self.__base = 0
    self.__height = 0
    self.setHeight(height)
    self.setBase(base)

  def getBase(self):
    return self.__base
  def setBase(self, new_base):
    self.__base = new_base
  def getHeight(self):
    return self.__height
  def setHeight(self, new_height):
    self.__height = new_height
  def area(self):
    return 0.5* self.__height * self.__base

t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())

t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
#task 4 done

class Team:
  def __init__(self, *args):
    self.__country_player_dict = {}

    if len(args) == 1 :
      self.__new_name = args[0]
      for country in args:
        self.__country_player_dict[country] = []

  def setName(self, new_name):
    self.__new_name = new_name
    self.__country_player_dict[new_name] = []

  def addPlayer(self, location):
    self.__country_player_dict[self.__new_name].append(location.player)

  def printDetail(self):
    print(f"=====================\nTeam: {self.__new_name}\nList of players:")
    for player_list in self.__country_player_dict.values():
      print(player_list)
    print("=====================")
  

class Player:
  def __init__(self, new_player):
    self.player = new_player

b = Team()
b.setName('Bangladesh')
mashrafi = Player("Mashrafi")
b.addPlayer(mashrafi)
tamim = Player("Tamim")
b.addPlayer(tamim)
b.printDetail()
a = Team("Australia")
ponting = Player("Ponting")
a.addPlayer(ponting)
lee = Player("Lee")
a.addPlayer(lee)
a.printDetail()
```

```python
#task 8 done (correct but complexicity needs to be reduced)

class Library:
  def __init__(self, location, book_dict):

    self.__location = None
    self.setLocation(location)

    self.__borrowerName_booksBorrowed_dict = {}

    self.__available_book_dict = None
    self.setBookDict(book_dict)

  def details(self):
    print(f"{self.__location} Library Details\nBorrower Details: \n{self.__borrowerName_booksBorrowed_dict}\nBooks availability:\n{self.__available_book_dict}")

  def setLocation(self, loc):
    self.__location = loc

  def setBookDict(self, dict_):
    self.__available_book_dict = dict_

  def getBookDict(self):
    return self.__available_book_dict

  def getLocation(self):
    return self.__location

  def getBorrowerNameBooksBorrowedDict(self):
    return self.__borrowerName_booksBorrowed_dict

class Reader:
  def __init__(self, reader_name):
    self.reader_name = reader_name
    self.borrowed_books_count = 0
    self.borrowedbooktype_count_dict = {}

  def borrow(self, loc, *books):

    for book in books:

      if self.borrowed_books_count == 0:

        if loc.getBookDict()[book] == 0:

          print(f"{book} books are not available at the moment.")

        else:

          loc.getBorrowerNameBooksBorrowedDict()[self.reader_name] = 1
          loc.getBookDict()[book] -= 1
          self.borrowed_books_count += 1
          print(f"{book} book is borrowed successfully.")

          if book in self.borrowedbooktype_count_dict.keys():
            self.borrowedbooktype_count_dict[book] += 1

          else:
            self.borrowedbooktype_count_dict[book] = 1

      elif self.borrowed_books_count < 5:

        if loc.getBookDict()[book] == 0:
          print(f"{book} books are not available at the moment.")

        else:
          loc.getBookDict()[book] -= 1
          self.borrowed_books_count += 1

          print(f"{book} book is borrowed successfully.")
          loc.getBorrowerNameBooksBorrowedDict()[self.reader_name] += 1

          if book in self.borrowedbooktype_count_dict.keys():
            self.borrowedbooktype_count_dict[book] += 1

          else:
            self.borrowedbooktype_count_dict[book] = 1

      else:
        print("You cannot borrow more than 5 books.")
        break

  def readerInfo(self, book_type = None):

    if book_type == None:
      print(f"{self.reader_name}, you have borrowed {self.borrowed_books_count} book(s) with you.")
      for booktype, bookcount in self.borrowedbooktype_count_dict.items():
        print(f"Books on {booktype}: {bookcount}")
    else:
      print(f"{self.reader_name}, you have {self.borrowedbooktype_count_dict[book_type]} on {book_type}(s) with you.")


  def preaderInfo(self):
    print(f"{self.reader_name}, you have {self.borrowed_books_count} book(s) with you.")

    for booktype, bookcount in self.borrowedbooktype_count_dict.items():
      print(f"Books on {booktype}: {self.borrowed_books_count}")

L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.preaderInfo()
print("8----------------------")
L1.details()
```
